```python
from Bio.PDB import PDBParser
from Bio.PDB.Polypeptide import standard_aa_names
from Bio.PDB.Polypeptide import protein_letters_3to1


parser=PDBParser()
structure =  parser.get_structure('tumor_necrosis','1a8m.pdb')
sequence = [] # Initialize an empty list to store the sequence

for model in structure:
  print('model',model.id)
  for chain in model:
    if chain.id=='A':
      for residue in chain:
        if residue.resname in standard_aa_names:
          res_id = residue.id[1]
          if res_id == 10 and residue.resname == 'HIS':
            sequence.append('ALA')
          else:
            sequence.append(residue.resname)

sequence_1_letter = ''.join([protein_letters_3to1.get(res,'X') for res in sequence])
print('amino acid sequence (1 letter codes):')
print(sequence_1_letter)
print('mutated seq.[residue 10 changed to ALA]:')
print(sequence_1_letter)
print('mut',residue.resname,residue.id)
```
